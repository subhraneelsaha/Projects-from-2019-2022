```python
from google.colab.patches import cv2_imshow
```

The only library we need for converting an image into a pencil sketch with Python is an OpenCV library in Python

```python
import cv2
```

Now the next thing to do is to read the image:

```python
img = cv2.imread("dog.jpg")
```

```python
cv2_imshow(img)
```

Now after reading the image, we will create a new image by converting the original image to greyscale:

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img_gray)
```

Now the next step is to invert the new grayscale image:

```python
img_invert = cv2.bitwise_not(img_gray)
```

```python
cv2_imshow(img_invert)
```

Now the next step in the process is to blur the image by using the Gaussian Function in OpenCV:

```python
img_smoothing = cv2.GaussianBlur(img, (21, 21), sigmaX=0, sigmaY=0)
img_smoothing_gray = cv2.GaussianBlur(img_gray, (21, 21), sigmaX=0, sigmaY=0)
img_smoothing_gray_invert = cv2.GaussianBlur(img_invert, (21, 21), sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
cv2_imshow(img_smoothing_gray)
cv2_imshow(img_smoothing_gray_invert)
```

Then the final step is to invert the blurred image, then we can easily convert the image into a pencil sketch:

```python
inverted_blurred = 255 - img_smoothing_gray_invert
pencil_sketch = cv2.divide(img_gray, inverted_blurred, scale=256.0)
cv2_imshow(img)
cv2_imshow(pencil_sketch)
```
